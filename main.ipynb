# Scaling, Train-Test Split, and Vanilla Models

```python
# Rescaling the dataset such that the mean is 0 and the standard deviation is 1 for every feature
my_scaler = StandardScaler(with_mean = 0, with_std = 1)
my_scaler.fit(dataset[features])
X_full = my_scaler.transform(dataset[features])
y_full = dataset['label']

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_full, y_full, test_size=0.2, random_state=42)

# Function that returns the report for the model on the holdout dataset
def check_accuracy(model, X_test, y_test):
    y_pred = model.predict(X_test)
    print('Name:', model.__class__.__name__)
    print('Parameters:', model.get_params())
    print('Accuracy: ', classification_report(y_test, y_pred))

# Function that returns plots the confusion matrix for the model on the holdout dataset
def plot_confusion_matrix(model, X_test, y_test):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(4, 4))
    sns.heatmap(cm, annot=True, fmt='.0f', cmap='OrRd', xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
    plt.title('Confusion Matrix', fontsize = 10)
    plt.xlabel('Predicted Label', fontsize = 10)
    plt.ylabel('True Label', fontsize = 10)
    plt.show()
```

KNN Classifier

```python
clf = KNeighborsClassifier()

clf.fit(X_train, y_train)

check_accuracy(clf, X_test, y_test)
plot_confusion_matrix(clf, X_test, y_test)
```

Vanilla Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()

clf.fit(X_train, y_train)

check_accuracy(clf, X_test, y_test)
plot_confusion_matrix(clf, X_test, y_test)
```

# Investigating KNN Hyperparameters

In this section, I have looked how changing some hyperparameters in the KNN model affect the accuracy "in vacuum"(i.e. without any other hyperparameters being changed). 

Let's first look how the <i>n_neighbors</i> hyperparameter affects the accuracy of the model. We plot the cross validation accuracy against the number of neighbors.

```python
n_neighbors = np.arange(1, 100, 2)
cv_scores_mean = []
cv_scores_std = []

for k in n_neighbors:
    clf = KNeighborsClassifier(n_neighbors = k)
    cv_score = cross_val_score(clf, X_full, y_full, cv=10)
    cv_scores_mean.append(np.mean(cv_score))
    cv_scores_std.append(np.std(cv_score))

plt.errorbar(n_neighbors, cv_scores_mean, yerr=cv_scores_std, fmt='o', capsize=5)
sns.lineplot(x = n_neighbors, y = cv_scores_mean)
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.grid(True)
plt.title('KNN against number of neighbors')
plt.show()
```

We see that the accuracy is highest when <i>n_neighbors</i> is in the range of 20-50. After than, the accuracy slowly decreases. We can also notice that the variance of the accuracy is pretty high. 

I plot the same graph for different distace metrics. 

```python
def knn_line_plot(metric):
    n_neighbors = np.arange(1, 100, 2)
    cv_scores_mean = []
    cv_scores_std = []

    for k in n_neighbors:
        clf = KNeighborsClassifier(n_neighbors = k, metric = metric)
        cv_score = cross_val_score(clf, X_full, y_full, cv=10)
        cv_scores_mean.append(np.mean(cv_score))
        cv_scores_std.append(np.std(cv_score))

    sns.lineplot(x = n_neighbors, y = cv_scores_mean, label = metric)

metrics = ['euclidean', 'manhattan', 'chebyshev', 'cosine']

for metric in metrics:
    knn_line_plot(metric)

plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.grid(True)
plt.title('KNN against number of neighbors')
plt.show()
```

Manhattan distance seems to perform the best. 

Earier we saw that the data features are highly correlated. We can use PCA to reduce the dimensionality of the data. Let's see how the number of components affects the accuracy of the model. 

```python
n_components = np.arange(1, 30, 2)
cv_scores_mean = []
cv_scores_std = []

for k in n_components:
    pca = PCA(n_components = k)
    X = pca.fit_transform(X_full)
    y = y_full

    clf = KNeighborsClassifier(n_neighbors = 30)

    cv_score = cross_val_score(clf, X, y, cv=5)
    cv_scores_mean.append(np.mean(cv_score))
    cv_scores_std.append(np.std(cv_score))

clf = KNeighborsClassifier(n_neighbors = 30)
knn_cv = cross_val_score(clf, X_full, y_full, cv=5)
plt.hlines(y = np.mean(knn_cv), xmin = 0, xmax = 30, colors = 'red', linestyles = 'dashed', label = 'KNN without PCA')
plt.errorbar([1], [np.mean(knn_cv)], yerr=[np.std(knn_cv)], fmt='o', capsize=5)

plt.errorbar(n_components, cv_scores_mean, yerr=cv_scores_std, fmt='o', capsize=5)
sns.lineplot(x = n_components, y = cv_scores_mean, label = 'KNN with 30 neighbors')

plt.xlabel('Number of components')
plt.ylabel('Accuracy')
plt.grid(True)
plt.title('KNN with 30 neighbors against PCA components')
plt.legend()
plt.show()
```

We can see that PCA does not imrpove the accuracy of the KNN model. 

We now check if using an ensemble method like bagging improves the accuracy of the model. 

```python
n_neighbors = np.arange(1, 30, 1)
bagging_cv_scores_mean = []
bagging_cv_scores_std = []
cv_scores_mean = []
cv_scores_std = []

for k in n_neighbors:
    bagging_clf = BaggingClassifier(estimator=KNeighborsClassifier(n_neighbors = k), n_estimators = 100, max_samples = 0.5, max_features = 0.8)
    bagging_cv_score = cross_val_score(bagging_clf, X, y, cv=10)
    bagging_cv_scores_mean.append(np.mean(cv_score))
    bagging_cv_scores_std.append(np.std(cv_score))

    knn_clf = KNeighborsClassifier(n_neighbors = k)
    cv_score = cross_val_score(knn_clf, X, y, cv=10)
    cv_scores_mean.append(np.mean(cv_score))
    cv_scores_std.append(np.std(cv_score))

sns.lineplot(x = n_neighbors, y = cv_scores_mean, label = 'KNN')
sns.lineplot(x = n_neighbors, y = bagging_cv_scores_mean, label = 'Bagging+KNN')

plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy Variance')

plt.grid(True)
plt.title('KNN vs Bagging+KNN comparison', fontsize = 15)
plt.legend()
plt.show()
```

So bagging, only gives us a slight improvement on small number of neighbors. This is not surprising since choosing the high nubmer of neighbors gives the same ensembling effect. 

We try to use Polynomial Transformation (<a href="https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html#sklearn.preprocessing.PolynomialFeatures">link</a>) to see if it improves the accuracy of the model. However, it does not.

```python
from sklearn.preprocessing import PolynomialFeatures


for degree in np.arange(1, 5):
    poly = PolynomialFeatures(degree=degree)
    poly.fit(X_full)
    X_full_poly = poly.transform(X_full)

    n_neighbors = np.arange(1, 100, 2)
    cv_scores_mean = []
    cv_scores_std = []

    for k in n_neighbors:
        clf = KNeighborsClassifier(n_neighbors = k)
        cv_score = cross_val_score(clf, X_full_poly, y_full, cv=10)
        cv_scores_mean.append(np.mean(cv_score))
        cv_scores_std.append(np.std(cv_score))

    sns.lineplot(x = n_neighbors, y = cv_scores_mean, label = 'KNN with Poly Degree ' + str(degree))
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.grid(True)
plt.title('KNN against number of neighbors')
plt.show()
```

# Investigating Random Forest Hyperparameters

In this section, I have looked how changing some hyperparameters in the KNN model affect the accuracy "in vacuum"(i.e. without any other hyperparameters being changed). 

Let's first look how the <i>max_depth</i> hyperparameter affects the accuracy of the model. We plot the cross validation accuracy against the maximum depth of the trees.

```python
max_depth_values = np.arange(1, 300, 20)
cv_scores_mean = []
cv_scores_std = []

for k in max_depth_values:
    print(f'k = {k}', sep = ' ')
    clf = RandomForestClassifier(max_depth = k)
    cv_score = cross_val_score(clf, X_full, y_full, cv=10)
    cv_scores_mean.append(np.mean(cv_score))
    cv_scores_std.append(np.std(cv_score))

plt.errorbar(max_depth_values, cv_scores_mean, yerr=cv_scores_std, fmt='o', capsize=5)
sns.lineplot(x = max_depth_values, y = cv_scores_mean)
plt.xlabel('Maximum depth')
plt.ylabel('Accuracy')
plt.grid(True)
plt.title('Random Forest Classifier accuracy against number of neighbors')
plt.show()
```

We can see that there is no improvement in accuracy after the depth of ~30. This is likely due to the fact that maximum possible depth is reach at that point.

We now look at the <i>n_estimators</i> hyperparameter. We fix <i>max_depth</i> at 50. 

```python
n_estimators_values = np.arange(1, 1000, 50)
cv_scores_mean = []
cv_scores_std = []

for k in n_estimators_values:
    print(f'k = {k}', sep = ' ')
    clf = RandomForestClassifier(n_estimators = k, max_depth = 50)
    cv_score = cross_val_score(clf, X_full, y_full, cv=10)
    cv_scores_mean.append(np.mean(cv_score))
    cv_scores_std.append(np.std(cv_score))

plt.errorbar(n_estimators_values, cv_scores_mean, yerr=cv_scores_std, fmt='o', capsize=5)
sns.lineplot(x = n_estimators_values, y = cv_scores_mean)
plt.xlabel('Maximum depth')
plt.ylabel('Accuracy')
plt.grid(True)
plt.title('Random Forest Classifier accuracy against number of estimators')
plt.show()
```

Again, we see that there is no improvement in accuracy after the number of estimators reaches ~100.

We check if PCA has any effect on the accuracy of the Random Forest model. We fix <i>n_estimators</i> at 200 and <i>max_depth</i> at 50.

```python
n_components = np.arange(1, 31, 2)
cv_scores_mean = []
cv_scores_std = []

for k in n_components:
    pca = PCA(n_components = k)
    X = pca.fit_transform(X_full)
    y = y_full

    clf = RandomForestClassifier(n_estimators= 100, max_depth = 50)

    cv_score = cross_val_score(clf, X, y, cv=5)
    cv_scores_mean.append(np.mean(cv_score))
    cv_scores_std.append(np.std(cv_score))

clf = RandomForestClassifier(n_estimators= 100, max_depth = 50)
rfc_cv = cross_val_score(clf, X_full, y_full, cv=5)
plt.hlines(y = np.mean(rfc_cv), xmin = 0, xmax = 30, colors = 'red', linestyles = 'dashed', label = 'RFC without PCA')
plt.errorbar([1], [np.mean(rfc_cv)], yerr=[np.std(rfc_cv)], fmt='o', capsize=5)

plt.errorbar(n_components, cv_scores_mean, yerr=cv_scores_std, fmt='o', capsize=5)
sns.lineplot(x = n_components, y = cv_scores_mean, label = 'PCA + RFC')

plt.xlabel('Number of components')
plt.ylabel('Accuracy')
plt.grid(True)
plt.title('RFC accuracy against the number of PCA components')
plt.legend()
plt.show()
```

# Hyperparameter Tuning

We use RandomizedSearchCV to find the best hyperparameters for our models. To prevent overfitting to the test set, we tune our models only on the training set. Then we evaluate the best model on the test set. This way, if there is overfitting, the the accuracy on the test set will be lower. However, we didn't see any overfitting in our models.

```python
def applyRFCwithHyperparameterTuning(data_train, labels_train, data_val, labels_val, title):
    param_grid = {
        'bootstrap': [True, False],
        'max_depth': [10, 20, 30, 40, 50, 60, 70, 80, 90],
        'min_samples_leaf': [1, 2, 4],
        'min_samples_split': [2, 5, 10],
        'n_estimators': [160, 320, 640, 1000]
    }
    
    clf = RandomizedSearchCV(RandomForestClassifier(), param_grid, n_iter = 100, refit = True, verbose = 3)
    clf.fit(data_train, labels_train)

    print('Best params', clf.best_params_)
    print(f'Train data accuracy: {clf.best_score_}')
    print(f'Validation data accuracy: {clf.score(data_val, labels_val)}')

    return clf

def applyKNNwithHyperparameterTuning(data_train, labels_train, data_val, labels_val, title):
    param_grid = {
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan', 'chebyshev', 'cityblock', 'cosine'],
        'n_neighbors': np.arange(1, 100, 2)
    }
    
    clf = RandomizedSearchCV(KNeighborsClassifier(), param_grid, n_iter = 100, refit = True, verbose = 3)
    clf.fit(data_train, labels_train)

    print('Best params: ', clf.best_params_)
    print(f'Train data accuracy: {clf.best_score_}')
    print(f'Validation data accuracy: {clf.score(data_val, labels_val)}')

    return clf
```

```python
rfc_hype_tuning = applyRFCwithHyperparameterTuning(X_train, y_train, X_test, y_test, 'Random Forest Classifier')
```

```python
knn_hype_tuning_with_scaling = applyKNNwithHyperparameterTuning(X_train, y_train, X_test, y_test, 'KNN')
```

Interestingly, KNN performs better on the unscaled dataset. This implies that the scales of the features are important for the KNN model. 

```python
X_unscaled = dataset[features]
y_unscaled = dataset['label']

X_train_unscaled, X_test_unscaled, y_train_unscaled, y_test_unscaled = train_test_split(X_unscaled, y_unscaled, test_size=0.2, random_state=42)

knn_hype_tuning_without_scaling = applyKNNwithHyperparameterTuning(X_train_unscaled, y_train_unscaled, X_test_unscaled, y_test_unscaled, 'KNN')
```

ROC curves typically feature true positive rate (TPR) on the Y axis, and false positive rate (FPR) on the X axis. This means that the top left corner of the plot is the “ideal” point - a FPR of zero, and a TPR of one. This is not very realistic, but it does mean that a larger area under the curve (AUC) is usually better. The “steepness” of ROC curves is also important, since it is ideal to maximize the TPR while minimizing the FPR. <a href = "https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html">link</a>

```python
from sklearn.metrics import roc_curve
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import RocCurveDisplay

def find_roc_curve(model, class_of_interest, x_test, y_test):
    y_score = model.predict_proba(X_test)

    label_binarizer = LabelBinarizer().fit(y_train)
    y_onehot_test = label_binarizer.transform(y_test)
    label_binarizer.transform([label])

    class_id = np.flatnonzero(label_binarizer.classes_ == class_of_interest)[0]

    RocCurveDisplay.from_predictions(
        y_onehot_test[:, class_id],
        y_score[:, class_id],
        name=f"{class_of_interest} vs the rest",
        color="darkorange",
    )
    plt.plot([0, 1], [0, 1], "k--", label="chance level (AUC = 0.5)")
    plt.axis("square")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title(f"One-vs-Rest ROC curves: {class_of_interest} vs Rest")
    plt.legend()
    plt.show()

```
